```python
"""
LoRA de Estilo Pré-treinada - Aqui usamos uma LoRA já treinada, em um pipeline base, para "fine-tunar" o modelo, ou seja, usamos um pequeno modelo específico para permitir que um estilo que não tá no modelo seja exportado no artefato.
"""
```

```python
import torch
from diffusers import DiffusionPipeline
from PIL import Image
```

```python
# Carregar o pipeline base

pipe = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
# Carregar os pesos da LoRA

lora_model_id = "dreamlike-art/dreamlike-diffusion-1.0"
pipe.load_lora_weights(lora_model_id, weight_name="dreamlike-art-sd-1.5.safetensors")

# Definir o prompt, incluindo a "trigger word"
# Trigger Word (Palavra de Gatilho): Note a palavra dreamlikeart. Quase toda LoRA precisa de uma palavra mágica no prompt para ser ativada. É como dizer ao modelo: "Use aquele treinamento especial que eu carreguei agora". Sem essa palavra, o efeito da LoRA pode ser fraco ou inexistente

prompt = "a portrait of a woman, dreamlikeart, trending on artstation"
```

```python
# Gerar a imagem
image = pipe(prompt, num_inference_steps=30, guidance_scale=7.5).images[0] 


image.save("teste_lora.png")
```
